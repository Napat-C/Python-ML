```python
# for quick calculation
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
age = [[50], [18], [35], [44], [29], [30], [60], [55], [25], [43]]
age = np.array(age)
normalization = scaler.fit_transform(age)
df = pd.DataFrame(normalization)
df.round(2)
```
