```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
df = pd.read_excel('example.xlsx')
df
```

```python
ohe = OneHotEncoder(sparse=False)
one_hot = pd.DataFrame(ohe.fit_transform(df[['Shape']]))
one_hot
```

```python
oe = OrdinalEncoder(categories=[['first','second','third'],
                                ['S','M','L','XL']])
ordinal = pd.DataFrame(oe.fit_transform(df[['Class','Size']]))
ordinal
```

```python
df = pd.concat([one_hot,ordinal],axis=1)
df
```

```python
df.columns = ['shape_circle','shape_oval','shape_square','Class','Size']
df
```

```python
df = df.drop('shape_square',axis=1)
df
```
